### 1) Mean / Median / Mode Imputation

when should we apply?
--> Mean / Median / Mode Imputation" has teh assumption that the data are missing completeley at Random (MCAR)
--> It is solved by replacing the NAN with the most frequent occurence of the variables

```python
df
```

```python
# read 3 colums from Titanic dataset
df = pd.read_csv('C:/Users/Kapil Wanaskar/Data Sets/titanic_train (1).csv',usecols=['age','fare','survived'])
df
```

```python
#percentage of missing values in each category
df.isnull().mean()
```

```python
#create a function which will impute NAN values
def impute_nan(df,variable,median):
# df,variable,median ~ are parameters
    df[variable+"_median"] = df[variable].fillna(median) # filling NAN values with Median
median = df.age.median()
median #median of age column
```

```python
impute_nan(df,'age',median) # applying the new funciton "impute_nan" to previous dataset of (age, fare survived)
df
# observe the NAN value in Age column is replaced with median = 28.0
```

```python
print(df['age'].std())  # Standard deviation of the orginal column "age"
print(df['age_median'].std()) # Standard deviation of the NEW column "age_median"
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_subplot(111) # defining axes
df['age'].plot(kind='kde', ax=ax) #plotted 1st, default colored, graph
# kde ~ kernel density estimator
df.age_median.plot(kind='kde', ax=ax, color = 'red') #plotted 2nd, red colored, graph
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best') #Naming / labeling the multi plots
```

### Advantages of Mean / Median / Mode Imputation

1) Easy to implement (Robust to Outliers)

2) Faster way to obtain the complete dataset

### Dis-Advantages of Mean / Median / Mode Imputation

1) Change / Distortion in the variance / Standard deviation  of the data

2) Impacts the Co-relation between dependent and independent variables
